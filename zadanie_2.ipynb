```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

import ml_module as ml

#загружаем датафрэйм
file_path = 'population_usa.csv'
df = ml.load_data(file_path)

ml.explore_data(df)
```

```python
ml.diagram(df, 'Year', 'Population')
```

```python
ml.heatmap(df)
```

```python
#препроцессинг
target_column = 'Population'
X, y, preprocessor = ml.preprocessing(df, target_column)

#разделение тестовое и тренировочной
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
model = ml.train_model(X_train, y_train)

#угадываем значение
y_pred = ml.predict(model, X_test)

#сторим метрики
ml.evaluate_model(y_test, y_pred)

#строим графики и не видим разницы между тестом и угаданным)
y_test_array = y_test.to_numpy()
ml.visualization(y_test_array, y_pred)
```
